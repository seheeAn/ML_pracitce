### Matplotlib

```python
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

##### Line Plots

```python
x = np.linspace(0,10,100) #0에서 10을 100개로 나눈다
plt.plot(x, np.sin(x))

plt.xlim([0,10])
plt.ylim([-2,2])

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

multi graph

```python
x=np.linspace(0,10,100)

plt.plot(x, np.sin(x), label='sin(x)', linestyle='', marker='o')
plt.plot(x, np.cos(x), label='cos(x)', linestyle='', marker='o')

plt.xlabel('x')
plt.ylabel('f(x)')

plt.legend(loc='lower left') #범례(legend) 추가 = lower left box
plt.show()
```

##### Scatter Plots

```python
rng = np.random.RandomState(123)
x = rng.normal(size = 500) 
y = rng.normal(size = 500) #make 500 random samples from normal distribution

plt.scatter(x,y)
plt.show()
```

##### Bar Plots

```python
means = [1,2,3]
stddevs = [0.2,0.4,0.5]
bar_labels = ['bar 1','bar 2','bar 3']

x_pos = list(range(len(bar_labels))) #list(range(3)) [0,1,2]
plt.bar(x_pos, means, yerr=stddevs) #yerr -> 막대 중간의 선으로 표시 

plt.show()
```

##### Histograms

```python
rng = np.random.RandomState(123)
x = rng.normal(0,20,1000) #mean=0, std.dev=20, sample size=1000

bins = np.arange(-100,100,5) #-100~100 5 steps

plt.hist(x,bins=bins)
plt.show()
```

```python
rng = np.random.RandomState(123)
x1 = rng.normal(0,20,1000) #mean=0, std.dev=20, sample size=1000
x2 = rng.normal(10,10,1000)

bins = np.arange(-100,100,5) #-100~100 5 steps

plt.hist(x1, label='x1', bins=bins, alpha=0.5)
plt.hist(x2, label='x2', bins=bins, alpha=0.5)
plt.legend(loc='upper right')

plt.show()
```

##### Subplots

```python
# import matplotlib.pyplot as plt

x = range(11)
y = range(11)

fig, ax = plt.subplots(nrows=2, ncols=3, 
                       sharex=True, sharey=True)

for row in ax:
    for col in row:
        col.plot(x,y)
        
plt.show()
```

##### Other functions of plotting

colors and markers

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),color='orange',marker='^',linestyle='')
plt.show()
```

saving plots

```python
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x))

plt.savefig('myplot.png', dpi=300) #save in file
plt.savefig('myplot.pdf')

plt.show()
```
